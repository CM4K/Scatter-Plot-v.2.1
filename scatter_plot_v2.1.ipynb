```python
import os
import time
import ffmpeg
import subprocess
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def savefig():
    #Set start, step, end frames, colors, area, dir, file, model:
    start = 0 #np.array(list(range(0, 4))
    step = 1
    end = 48 #max limit
    size = 80
    
    #set = np.random.normal(80)
    set = np.random.normal(start,step,end) #.uniform(start,step,end).rand(end)

    alpha = 0.4
    x = np.cumsum(set) #.cumsum #random.normal #rand #randint.exponential
    y = np.cumsum(set)
    colors = np.cumsum(set)
    area = np.pi * (8 * np.cumsum(set)) ** 4 #0 to 15 point radii - pattern 26 ** 4
    
    preset = 'Spectral'
    #Spectral,viridis,seismic,twilight,copper,winter,cool,cividis,bone,ocean,flag,prism,
    #binary,terrain,plasma,gist_earth,spring,summer,jet,inferno,gnuplot
    
    dir = 'output'
    full_dir = 'D:\Anaconda3\envs\plot\output'
    if not os.path.exists(dir): os.makedirs(dir)
    
    name = '\output'
    id = str({0})
    ext = '.png'
    type = '\%d'
    format = '.mp4'
    file = id + ext
    files = os.listdir(full_dir)
    dir_file = os.path.join(dir,file)
    
    for id, file in enumerate(str(len(files))):
        #for start in range(end): start += 1
        plt.figure(figsize=(28, 18)) #14, 6
        plt.scatter(x, y, s=area, c=colors, alpha=alpha, cmap=preset) #label="Luck" - #char =  u"✹"
        plt.axis('off') #plt.colorbar()
        plt.tight_layout()
        
        if len(files) == 0:
            plt.savefig('{}/{}{}'.format(dir,id,ext),bbox_inches='tight',transparent=True,pad_inches=0,dpi=100)
            plt.close()
            
            print('--> Output '+full_dir+'\n')
            print('--> Frame '+str(len(files)+1))

        else:
            id = len(files)
            plt.savefig('{}/{}{}'.format(dir,id,ext),bbox_inches='tight',transparent=True,pad_inches=0,dpi=100)
            plt.close() #plt.gcf() #plt.clf()
            
            if len(files) > end: 
                print('--> You must delete or remove output files before running this code again.') 
                break
                
            else: print('--> Frame '+str(len(files)+1))
            
        if len(files) >= end:
            video = 'ffmpeg -r 4 -i "'+full_dir+type+ext+'" -vcodec libx264 -crf 12 "'+full_dir+name+format+'"'
            subprocess.run(video, shell=True)
            break

        else:
            savefig()
            break
            
savefig()
```
